```python
import numpy as np

# 1. Append in NumPy
# np.append() is used to add values to an array.
# It creates a new array with the added elements.

arr = np.array([1, 2, 4])
# new = np.append(arr, [7, 8])
# print(new)

# Multi-Dimensional Example

# append row
# arr = np.array([[1, 2], [4, 5]])
# new = np.append(arr, [[7, 8]], axis = 0)
# print(new)

# Append a new column [[5], [6]]
# arr = np.array([[1, 2], [4, 5]])
# new = np.append(arr,[[7],[9]],axis=1)
# print(new)

# 2. Concatenate Arrays
# np.concatenate() joins arrays along existing axis.
# 2D Example - Along Rows (axis=0)
x = np.array([[1, 2], [3, 4]])
# y = np.array([[8, 9]])
# res = np.concatenate((x, y), axis=0)
# print(res)

# Concatenate Along Columns (axis=1)
# Note: Dono arrays mein number of rows (2) same honi chahiye
y_col = np.array([[8], [9]])
res_col = np.concatenate((x, y_col), axis=1)
print(res_col)
```
